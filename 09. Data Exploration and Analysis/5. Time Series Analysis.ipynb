# Time Series Analysis

```python
import pandas as pd
```

```python
aliens = pd.read_csv(r"C:\Users\alexf\OneDrive\Documents\Tutorials\Pandas Folder\EDA\alien_sightings.csv")
aliens
```

```python
aliens.info()
```

```python
aliens['datetime'] = pd.to_datetime(aliens['datetime'])
```

```python
aliens['duration (seconds)'] = aliens['duration (seconds)'].str.replace("`", "")
```

```python
aliens['duration (seconds)'] = pd.to_numeric(aliens['duration (seconds)'])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (12,6))
plt.scatter(aliens['datetime'],aliens['duration (seconds)'] )

plt.show()
```

```python
aliens.head()
```

```python
aliens.set_index('datetime',inplace = True)
```

```python
aliens.head()
```

```python
aliens.resample('M').count()
```

```python
monthly_counts = aliens.resample('M').count()
```

```python
plt.plot(monthly_counts.index, monthly_counts['shape'])
plt.show()
```

```python
monthly_counts_year = aliens.resample('Y').count()
```

```python
plt.plot(monthly_counts_year.index, monthly_counts_year['shape'])
plt.show()
```

```python
aliens.tail()
```

```python
sightings_by_country = aliens.groupby('country').resample('M').size().unstack(level=0)
```

```python
sightings_by_country.plot(figsize=(12,6))
```
