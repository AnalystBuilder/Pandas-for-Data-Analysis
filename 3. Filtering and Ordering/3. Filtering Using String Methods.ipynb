# Filtering Using String Methods

```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\alexf\OneDrive\Documents\Tutorials\Pandas Folder\Filtering Lesson\laptop_prices.csv',
            encoding = 'ISO-8859-1')

pd.set_option('display.max.rows',1350)
pd.set_option('display.max.columns',15)
```

```python
df[(df['Product'].str.contains('MacBook')) & (df['Memory'].str.contains('SSD'))]
```

```python
df[df['Product'].str.startswith('Mac')]
```

```python
df[df['Product'].str.endswith('14')]
```

```python
df[df['Price_usd'].duplicated()]
```

```python
df[df['Product'].str.match('Mac')]
```

```python
df[df['Product'].str.match('Mac.*Air')]
```
