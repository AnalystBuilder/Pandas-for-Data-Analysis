# Visualizations with Seaborn

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
housing = pd.read_csv(r"C:\Users\alexf\OneDrive\Documents\Tutorials\Pandas Folder\Data Visualization\real_estate.csv"
                 , encoding = "ISO-8859-1")
housing
```

```python
housing_filtered = housing[housing['state'].isin(['Massachusetts','Puerto Rico', 'New York'])]
```

```python
sns.barplot(housing_filtered, x = 'state',y = 'price', estimator = 'mean')
plt.plot()
```

```python
housing_agg = housing_filtered.groupby('state')['price'].mean().reset_index()
```

```python
housing_agg.plot(x = 'state', y='price', kind = 'bar')
```

```python
sns.set_style('darkgrid')
```

```python
sns.barplot(housing_filtered, x = 'state',y = 'price', estimator = 'mean')
plt.plot()
```

```python
sns.set_palette('Reds')
```

```python
sns.barplot(housing_filtered, x = 'state',y = 'price', estimator = 'mean')

plt.title('Average State by Price')

plt.plot()
```

```python
sns.set_palette('deep')

housing_filt = housing[housing['house_size'] < 10000]

sns.regplot(data = housing_filt, x = 'house_size', y = 'price', line_kws={'color':'red'})

plt.ticklabel_format(style='plain', axis = 'y')
plt.show()
```
