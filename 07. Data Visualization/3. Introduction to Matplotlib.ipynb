# Introduction to Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [5,10,20]

plt.plot(x,y, label = 'Showers', color = 'green', marker = '^', markersize = 15, linestyle = '--', markeredgecolor = 'blue')

plt.title('Matplotlib Viz', fontdict = {'fontname': 'Times New Roman', 'fontsize': 30})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([1,2,3])
plt.yticks([0,5,10,15,20])

plt.legend()

plt.show()
```

```python
x = [1,2,3]
y = [5,10,20]

plt.plot(x,y, label = 'Showers', color = 'green', marker = '^', markersize = 15, linestyle = '--', markeredgecolor = 'blue')

z = [1,5,10]
plt.plot(x,z, label = 'Baths', color = 'red', marker = '.', markersize = 20, linestyle = '--', markeredgecolor = 'green')


plt.title('Matplotlib Viz', fontdict = {'fontname': 'Times New Roman', 'fontsize': 30})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([1,2,3])
plt.yticks([0,5,10,15,20])

plt.legend()

plt.show()
```

```python
x = [1,2,3]
y = [5,10,20]

plt.figure(figsize = (20,5))

plt.plot(x,y, label = 'Showers', color = 'green', marker = '^', markersize = 15, linestyle = '--', markeredgecolor = 'blue')

z = [1,5,10]
plt.plot(x,z, label = 'Baths', color = 'red', marker = '.', markersize = 20, linestyle = '--', markeredgecolor = 'green')


plt.title('Matplotlib Viz', fontdict = {'fontname': 'Times New Roman', 'fontsize': 30})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([1,2,3])
plt.yticks([0,5,10,15,20])

plt.legend()


plt.savefig(r"C:\Users\alexf\OneDrive\Documents\Tutorials\Pandas Folder\Data Visualization\matplotlib.png")

plt.show()
```
