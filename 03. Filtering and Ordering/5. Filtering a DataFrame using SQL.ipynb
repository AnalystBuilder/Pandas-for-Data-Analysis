# Filtering a DataFrame using SQL

```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\alexf\OneDrive\Documents\Tutorials\Pandas Folder\Filtering Lesson\laptop_prices.csv',
            encoding = 'ISO-8859-1')

pd.set_option('display.max.rows',1350)
pd.set_option('display.max.columns',15)
```

```python
df.query("Price_usd > 1200")
```

```python
df.query("Company == 'Apple' or Company == 'Dell'")
```

```python
df.query("Company == 'Apple' or Price_usd > 1500")
```

```python
import pandasql as psql
```

```python
psql.sqldf("SELECT * FROM df WHERE Company = 'Apple'")
```

```python
psql.sqldf("SELECT Company FROM df WHERE Company = 'Apple' AND Inches > 14")
```

```python
psql.sqldf("SELECT Company, Product FROM df WHERE Product LIKE '%MacBook%'")
```

```python
psql.sqldf("SELECT * FROM df ORDER BY Price_usd DESC ")
```
