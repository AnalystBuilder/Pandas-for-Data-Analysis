# Transform with Group By

```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\alexf\OneDrive\Documents\Tutorials\Pandas Folder\Group By Lesson\banking.csv'
                , encoding = 'ISO-8859-1')

pd.set_option('display.max.rows',46000)
```

```python
df_banking = df[:500]
```

```python
df_banking['average_balance_per_job'] = df_banking.groupby('job')['outstanding_balance'].transform('mean')
```

```python
df_banking['above_below_average'] = df_banking.groupby('job')['outstanding_balance'].transform(lambda x: x - x.mean())
```

```python
df_banking
```

```python
df_banking['average_balance_with_loan'] = df_banking[df_banking['housing_loan'] == 'yes'].groupby('job')['outstanding_balance'].transform('mean')
```
