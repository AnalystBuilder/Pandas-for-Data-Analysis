# Grouping and Visualizing Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
insurance = pd.read_csv(r"C:\Users\alexf\OneDrive\Documents\Tutorials\Pandas Folder\EDA\health_insurance.csv")
insurance
```

```python
grouped_by_region = insurance.groupby('region').mean()['charges']
```

```python
sns.barplot(x = grouped_by_region.index, y = grouped_by_region.values)
plt.xlabel("Region")
plt.ylabel("Average Yearly Charges")
plt.title("Average Charges per Region")

plt.show()
```

```python
sns.histplot( data = insurance, x = 'charges',hue = 'smoker')
plt.xlabel("Yearly Healthcare Cost")
plt.ylabel("Count of Individuals")
plt.title("Distrbution of Charges for Smokers and Non-Smokers")

plt.show()
```

```python
sns.histplot(data = insurance, x = 'bmi', y = 'charges', bins = 30)

plt.xlabel("BMI")
plt.ylabel("Charges Yearly")
plt.title("Charges vs BMI")

plt.show()
```

```python
sns.scatterplot(data = insurance, x = 'age', y = 'charges', hue = 'bmi', size = 'bmi',sizes = (20, 200), palette = 'coolwarm' )

plt.xlabel("Age")
plt.ylabel("Charges")
plt.title("Relationship between Age, Charges, and BMI")

plt.show()
```
